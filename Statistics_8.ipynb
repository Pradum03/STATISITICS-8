1.  Calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation 
of 5 using Python. Interpret the results.

```python
import scipy.stats as stats

# set the sample mean, standard deviation, sample size, and confidence level
sample_mean = 50
sample_std = 5
sample_size = 100
conf_level = 0.95

# calculate the standard error
std_error = sample_std / (sample_size ** 0.5)

# calculate the t-value for the given confidence level and degrees of freedom
t_value = stats.t.ppf((1 + conf_level) / 2, sample_size - 1)

# calculate the confidence interval
lower = sample_mean - t_value * std_error
upper = sample_mean + t_value * std_error

# print the results
print(f"95% confidence interval: ({lower:.2f}, {upper:.2f})")
```

4.  A study of the prevalence of smoking in a population of 500 individuals found that 60 individuals 
smoked. Use Python to calculate the 95% confidence interval for the true proportion of individuals in the 
population who smoke

```python

import numpy as np

# Sample size
n = 500

# Number of individuals who smoke
x = 60

# Sample proportion
p = x/n

# Z-score for a 95% confidence interval
z = 1.96

# Margin of error
moe = z * np.sqrt((p*(1-p))/n)

# Confidence interval
CI = (p - moe, p + moe)

print("95% Confidence Interval: ", CI)
```

To calculate the confidence interval for the true proportion of individuals in the population who smoke, we can use the formula:

CI = p ± zsqrt((p(1-p))/n)

where:

CI is the confidence interval
p is the sample proportion of individuals who smoke
z is the z-score for the desired confidence level (z=1.96 for a 95% confidence interval)
n is the sample size

5. Calculate the 90% confidence interval for a sample of data with a mean of 75 and a standard deviation 
of 12 using Python. Interpret the results.

```python
import numpy as np
import scipy.stats as stats

sm= 75
n=100  # assuming n=100 as it is not given
dof=n-1
sigma=12
ci=0.90

std_error = sigma/ (n ** 0.5)

# calculate the t-value for the given confidence level and degrees of freedom
t_value = stats.t.ppf((1 + ci) / 2, dof)

# calculate the confidence interval
lower = sm - t_value * std_error
upper = sm + t_value * std_error

# print the results
print(f"90% confidence interval: ({lower:.2f}, {upper:.2f})")
```

6. Use Python to plot the chi-square distribution with 10 degrees of freedom. Label the axes and shade the 
area corresponding to a chi-square statistic of 15

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chi2

# Define the degrees of freedom
df = 10

# Generate x values from 0 to 30
x = np.linspace(0, 30, 500)

# Create the chi-square distribution with 10 degrees of freedom
chi2_dist = chi2(df)

# Plot the chi-square distribution
plt.plot(x, chi2_dist.pdf(x), label="Chi-square PDF")

# Shade the area corresponding to a chi-square statistic of 15
x_shade = np.linspace(15, 30, 100)
plt.fill_between(x_shade, chi2_dist.pdf(x_shade), color="blue", alpha=0.3)

# Add labels and legend
plt.xlabel("Chi-square value")
plt.ylabel("Probability density")
plt.legend()

# Show the plot
plt.show()

```

7. A random sample of 1000 people was asked if they preferred Coke or Pepsi. Of the sample, 520 
preferred Coke. Calculate a 99% confidence interval for the true proportion of people in the population who 
prefer Coke.

```python
from scipy.stats import norm
import numpy as np

# Sample size
n = 1000

# Number of individuals who smoke
x = 520

# Sample proportion
p = x/n

# Z-score for a 99% confidence interval
confidence_level = 0.99
z_score = norm.ppf((1 + confidence_level) / 2)

# Margin of error
moe = z_score * np.sqrt((p*(1-p))/n)

# Confidence interval
CI = (p - moe, p + moe)

print("99% Confidence Interval: ", CI)
```

11.  A random sample of 30 people was selected from a population with an unknown mean and standard 
deviation. The sample mean was found to be 72 and the sample standard deviation was found to be 10. 
Conduct a hypothesis test to determine if the population mean is significantly different from 70. Use a 
significance level of 0.05

```python
# Ho = No difference in mean
# H1 = Difference in mean

import numpy as np
from  scipy.stats import t

n=30
dof= n-1
sample_mean=72
population_mean =70
sample_sd=10
alpha = 0.05
t_test = (sample_mean-population_mean)/((sample_sd)/n**0.5)
t_value = t.ppf(1 - alpha/2, dof)
if(t_test>t_value):
    print("Reject Null Hypothesis")
else:
    print("Accept Null Hypothesis")
```

To test whether the population mean is significantly different from 70 based on a sample of 30 people with sample mean 72 and sample standard deviation 10, we can perform a t-test.

The null hypothesis is that the population mean is equal to 70, while the alternative hypothesis is that the population mean is significantly different from 70. We will use a significance level of 0.05.

The test statistic for this t-test can be calculated as:

t = (sample mean - hypothesized population mean) / (sample standard deviation / sqrt(sample size))
t = (72 - 70) / (10 / sqrt(30))
t = 1.897

To find the p-value, we need to calculate the probability of observing a t-value as extreme or more extreme than 1.897 if the null hypothesis is true. We use a t-distribution with degrees of freedom equal to the sample size minus 1 (df = 30 - 1 = 29) to find the p-value. We will use a two-tailed test since we are testing whether the population mean is significantly different from 70.

Using a t-table or a t-distribution calculator with df = 29, we find that the p-value for t = 1.897 is approximately 0.067.

Since the p-value of 0.067 is greater than the significance level of 0.05, we fail to reject the null hypothesis. Therefore, we do not have sufficient evidence to conclude that the population mean is significantly different from 70 at the 0.05 level of significance
